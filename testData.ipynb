```python
import pandas as pd
dfna = pd.read_csv('league_matches_data_final_NA.csv')
dfkr = pd.read_csv('league_matches_data_final_KR.csv')

dfkr.head()
```

```python
len(dfkr['match_id'])
```

```python
dfkr = dfkr.drop_duplicates(subset='match_id')
len(dfkr['match_id'].unique())
dfna = dfna.drop_duplicates(subset='match_id')
len(dfna['match_id'].unique())
```

```python
df_combined = pd.concat([dfna, dfkr])
df_combined = df_combined.sample(frac=1).reset_index(drop=True)
df_combined.head()
```

```python
len(df_combined['match_id'].unique())
```

```python
df_combined.to_csv('combined_league_matches.csv', index=False)
```
